```python
#matplotlib 모듈 불러오기
import matplotlib.pyplot as plt
```

```python
# plot -> 기본 그래프 그리기
y = [2,4,6]
plt.plot(y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls='--')
plt.show()
```

```python
# -, -- , -. ,  :  라인스타일 ls = ' '
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls='-.')
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls=':')
plt.show()
```

```python
# Marker 찍기
plt.plot(y, marker='o')
plt.show()
```

```python
plt.plot(y, marker='v')
plt.show()
```

```python
plt.plot(y, marker='^')
plt.show()
```

```python
plt.plot(y, marker='s')
plt.show()
```

```python
plt.plot(y, marker='d')
plt.show()
```

```python
#Line weight 설정하기
plt.plot(y, lw=9)
plt.show()
```

```python
# Line color 변경하기
plt.plot(y, color='r')
plt.show()
```

```python
plt.plot(y, color='violet',lw = 3, marker = 'o', markerfacecolor = 'r', ms = 10, mec='b')
plt.show()
```

```python
#x 값 -> 0~6
#y 값 -> 1,4,5,8,9,5,3
x = [1,3,4,6,5,4,2]
y = [1,4,5,8,9,5,3]
plt.plot(y, color='violet',ls = '-.' ,lw = 3, marker = 'o', markerfacecolor = 'r', ms = 10, mec='b')
plt.show()
```
