```python
#1. pandas 모듈을 사용하여 파일 불러오기
#2. 불러온 파일을 score라는 변수명에 저장하기
#3. index를 '과목'으로 지정하기

import pandas as pd
```

```python
score = pd.read_csv("score.csv", encoding="euc-kr", index_col = "과목")
score
```

```python
#.sum() -> 각 열에 따라 합계가 계산되는 함수 defualt값 세로
score.sum()
```

```python
#.sum(axis=1) -> 각 행에 따라 합계가 계산됨
# axis가 생략되어 있거나 0값이라면 -> 위 아래를 더해진다.
# axis가 1의 값이라면 -> 좌 우로 더해진다
score.sum(axis=1)
```

```python
#전체행에서 처음부터 4반이라는 열까지 불러오기 그걸 행으로 나누기
#슬라이싱 복습필요 loc iloc 복습
score["평균2"]= score.loc[:,:"4반"].mean(axis=1)
score
```

```python
#컬럼 추가시 loc를 사용하면 새로운 행 추가
score.loc["반평균"] = score.mean()
```

```python
ages=[0,2,10,21,23,27,37,61,20,41,32,100]
bins=[0,15,25,35,60,99]
labels=["미성년자","청년","중년","장년","노년"]
```

```python
# cut() 사용법 
# pd.cut(데이터, 구분영역. labels = 카테고리)
cats = pd.cut(ages, bins, labels = labels)
```

```python
age_arr = pd.DataFrame(ages, columns = ['ages'])
age_arr
```

```python
#concat / merge -> 데이터 프레임들을 병합할 수 있는 함수
#concat -> 단순히 데이터를 연결하는 함수
#(기본적으로는 위/아래로 데이터 행을 연결)
#merge -> 두 데이터 프레임의 공통 열 혹은 인덱스를 기준으로 결합
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
result = pd.concat([df1, df2, df3])
result
```

```python
result1 = pd.concat([df1, df2, df3], keys=['x','y','z'])
result1
```

```python
df4 = pd.DataFrame({"B":['B2', 'B3', 'B6', 'B7'],
                   "D":['D2', 'D3', 'D6', 'D7'],
                   "F":['F2', 'F3', 'F6', 'F7']},index=[2,3,4,7])
```

```python
df4
```

```python
df1
```

```python
result = pd.concat([df1,df4], axis = 1)
result
```

```python
#join 속성 -> DataFrame을 병합할 때 어떤 기준으로 할 것인지 지정
result = pd.concat([df1,df4], axis = 1, join = 'inner')
result
```

```python
#.reindex() -> 어떠한 데이터를 기준으로 가져올것인지 지정
result = pd.concat([df1, df4], axis=1).reindex(df1.index)
result
```
