```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# ls line style
# lw line weight
# marker
# ms marker size
# mfc marker face color
# mec 마커 테두리 색상
# mew 마커 테두리 두께 
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls = '--', lw = 3, marker = 'o', mfc = 'r', ms = 10, mec='g', mew = 3)
plt.show()
```

```python
#그림 범위 지정하기
# xlim(), ylim() -> x축과 y축을 시작점을 지정해주는 함수
# xlim(시작, 끝)
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls = '--', lw = 3, marker = 'o', mfc = 'r', ms = 10, mec='g', mew = 3)
plt.xlim(-2, 10)
plt.ylim(-5, 12)
plt.grid()
plt.show() 
```

```python
# 축의 위치를 알려주는 표시 지점 -> 틱(tick)
# 틱에 쓰인 숫자나 문자 -> 틱라벨
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls = '--', lw = 3, marker = 'o', mfc = 'r', ms = 10, mec='g', mew = 3)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
# 그래프의 배경에 선을 그어주는 함수
plt.grid()
plt.show()
```

```python
# 여러개의 라인 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```

```python
# 그래프의 범례 만들기 -> label 속성, legend()
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend(fontsize=14) # -> 범례를 만들어주는 함수
plt.title('x,y,z Graph')
plt.show()
```

```python
x = range(7)
plt.bar(x,y)
# xticks 사용하여 tick label 지정
plt.xticks(x, ['월','화','수','목','금','토','일'])
plt.title('요일별 사망교통사고 건수')
# y축의 범위 500~650 사이로 지정
plt.ylim(500,650)
# x.label, y.label
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.show()
```

```python
#boolean 색인 사용하기 trure, false 다시 던져서 원하는 값만 불러오기 
car_to_car = data[data['사고유형_대분류']=='차대차']
```

```python
# 지역별로 그룹을 묶어서 사망자수를 한번에 그룹별로 더해줌
car_group = car_to_car[['사망자수','발생지시도']].groupby('발생지시도').sum()
```

```python
car_group
```

```python
x_label = car_group.index
```

```python
x_label
```

```python
range(0, len(x_label))
```

```python
x = range(0, len(x_label))
y = car_group.index
```

```python
y
```

```python
# figure() -> 그래프가 그려지는 영역
plt.figure(figsize = (20, 10))
#1. plot을 사용하여 그래프 그리기
plt.plot(car_group, ls = '--', marker = 'o', mfc='r', mec='black')
#2. xticks -> 지역명
plt.xticks(x, x_label)
#3, xlabel -> '지역'
plt.xlabel('지역')
#4. ylabel -> '사상자 수'
plt.ylabel('사상자 수')
#5. title -> '차vs차 교통사고의 사상자 수'
plt.title('차vs차 교통사고의 사상자 수')
plt.grid()
plt.show()
```

```python
time = data['발생년월일시']
```

```python
time
```

```python
time = time%100
```

```python
time
```

```python
labels = ["0~2", "3~5", "6~8", "9~11", "12~14", "15~17","18~20", "21~23"]
```

```python
bins = [-1, 2, 5, 8, 11, 14, 17, 20, 23]
labels = ["0~2", "3~5", "6~8", "9~11", "12~14", "15~17","18~20", "21~23"]
```

```python
result = pd.cut(time, bins, labels=labels)
```

```python
result
```

```python
rs = result.value_counts()
```

```python
rs
```

```python
#인덱스값 기준 sort_index()
#데이터값 기준 
rs.sort_index()
```

```python
# autopct -> pie 위에 표시 될 글자 형태를 알아서 %(백분율)로 변환
plt.figure(figsize = (5, 5))
plt.pie(rs, labels=labels, autopct='%1.1f%%') # -> 1.2, 26.2 소수점 한자리씩 나오게 하겠다.
plt.show()
```

```python
# autopct -> pie 위에 표시 될 글자 형태를 알아서 %(백분율)로 변환
plt.pie(rs, labels=labels, autopct='%1.1f%%', colors=['gray','green','red','yellow','blue']) # -> 1.2, 26.2 소수점 한자리씩 나오게 하겠다.
plt.show()
```
